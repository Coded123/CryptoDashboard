```python
exchange = ccxt.bittrex()

cols = ['date', 'open', 'high', 'low', 'close', 'vol']
dfs = []
#fetch data for TTM in the 1 day timeframe
eth_ohlcv = exchange.fetch_ohlcv('ETH/USDT', timeframe= '1d')
```

```python
df = pd.DataFrame(data = eth_ohlcv, columns = cols)

df['date'] = pd.to_datetime(df['date'], unit='ms')
```

```python
df
```

```python
#set the date as index
df.set_index('date', inplace=True)
```

```python
df
```

```python
#when was eth less than 2.5k in 2022
eth_lessthan2500 = (df['open'] < 2500) & (df.index > '2022-01-01')
```

```python
df.loc[eth_lessthan2500].head(50)
```

```python
#groupby month, mean for each month and round to 1 d.p for a clean look
Month_close_grp = df.groupby(pd.Grouper(level='date',freq='M'))['close'].mean()

Month_close_grp.round(1)
```

```python
#month grouping
months_grp = df.groupby(pd.Grouper(level='date',freq='M'))
```

```python
#shows the mean and median of the close values for each month
months_grp['close'].agg(['mean', 'median'])
```

```python
fig = go.Figure(data=[go.Candlestick(x=df.index,
                open=df['open'],
                high=df['high'],
                low=df['low'],
                close=df['close'])
                ])


fig.update_layout(
    title='ETH 1h for the TTM',
    yaxis_title='ETH Price ($)',
    shapes = [dict(
        x0='2021-03-19', x1='2021-03-19', y0=0, y1=1, xref='x', yref='paper',
        line_width=2)],
    annotations=[dict(
        x='2021-03-19', y=0.05, xref='x', yref='paper',
        )]
)

fig.show()
```

```python
#initially months_grp worked by itself then stopped working so now doing the mean in order to create candlestick
months_grp_mean = months_grp.mean()
```

```python
fig = go.Figure(data=[go.Candlestick(x=months_grp_mean.index,
                open=df['open'],
                high=df['high'],
                low=df['low'],
                close=df['close'])
                ])


fig.show()
```
